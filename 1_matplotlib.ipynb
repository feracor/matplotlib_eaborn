#Pyplot básico

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11) #Genera 11 números uniformemente distribuidos entre 0 y 5
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) #Genera un gráfico a partir de dos variables
plt.show() #Visualiza el gráfico
```

```python
plt.plot(x,y, 'ro' ) #Gráfico con puntos rojos
plt.show() #Visualiza el gráfico
```

```python
plt.plot(x,y, 'bx' ) #Gráfico con 'X' azules
plt.show() #Visualiza el gráfico
```

```python
plt.plot(x,y, 'rs-' ) #Gráfico con cuadrados rojos y línea continua
plt.show() #Visualiza el gráfico
```

```python
plt.plot(x,y, 'yD:' ) #Gráfico con diamantes amarillos y puntos consecutivos
plt.show() #Visualiza el gráfico
```

```python
plt.hist(x) #Histograma variable 'x'
plt.show() #Visualiza el gráfico
```

```python
plt.pie(x) #Diagrama de pie variable 'x'
plt.show() #Visualiza el gráfico
```

```python
plt.scatter(x,y) #Diagrama scatter
plt.show() #Visualiza el gráfico
```

```python
plt.boxplot(x) #Boxplot - Diagrama de caja de variable 'x'
plt.show() #Visualiza el gráfico
```

```python
plt.subplot(2,1,1) #Visualiza en 2 filas, 1 columna, index 1
plt.plot(x,y, 'gD:') #Gráfica de líneas, primeros datos
plt.plot(y,x, 'rx-') #Gráfica de líneas, segundos datos agregados
plt.subplot(2,1,2) #Visualiza en 2 fila, 1 columna, index 2
plt.hist(y) #Histrograma
plt.show() #Mostrar gráficas
```

#Subplots (orientado a objetos)

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b') #Posición 0
axes[1].plot(y,x,'r') #Posición 1
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2) #Declaración como tupla
ax1.plot(x,y,'b') #Posición 0
ax2.plot(y,x,'r') #Posición 1
```

```python
fig,axes = plt.subplots(2, 4) #2 filas, 4columnas
axes[0,0].plot(x,np.cos(x),'b') #Posición 0,0
axes[0,1].plot(x,np.sin(x),'g') #Posición 0,1
axes[0,2].plot(x,np.tan(x),'r') #Posición 0,2
axes[0,3].plot(x,np.arccos(x),'g') #Posición 0,3
```

```python
fig,((ax1, ax2, ax3, ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2, 4) #Se deben definir todas las 8 salidas
ax1.plot(x,np.cos(x),'b') #Posición 0,0
ax2.plot(x,np.sin(x),'g') #Posición 0,1
ax3.plot(x,np.tan(x),'r') #Posición 0,2
ax4.plot(x,np.arccos(x),'g') #Posición 0,3

ax5.plot(x,np.arccosh(x),'y') #Posición 1,0
ax6.plot(x,np.arcsin(x),'r') #Posición 1,1
ax7.plot(x,np.arctan(x),'g') #Posición 1,2
ax8.plot(x,np.arccos(x)**2,'b') #Posición 1,3

fig.tight_layout() #Distribuye uniformemente
```

#Leyendas, etiquetas, títulos, tamaño

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5)) #Propiedades tablero
axes[0].plot(x,y, label='$sin(x)$') #'label' define leyenda
axes[0].set_title('Relación X - Y') #Título gráfica 1
axes[0].set_xlabel('X') #Eje 'x' gráfica 1
axes[0].set_ylabel('Y') #Eje 'y' gráfica 1
axes[0].legend() #Muestra la leyenda

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X') #Título gráfica 2
axes[1].set_xlabel('Y') #Eje 'x' gráfica 2
axes[1].set_ylabel('X') #Eje 'y' gráfica 2
```

```python
#Sin orientación a objetos
plt.plot(x, y, label='sin(x)')
plt.title('El título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(0.1,0.2)) #bbox, posición de leyenda por coordenada
```

```python
#Sin orientación a objetos
plt.plot(x, y, label='sin(x)')
plt.title('El título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right') #Ubicación de leyenda
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
```

```python
#Customización con estilo matlab
fig, axes = plt.subplots(figsize=(4,4))

axes.plot(x, x+1, 'r--')
axes.plot(x, x+2, 'y-')
axes.plot(x, x+3, 'b-.')
axes.plot(x, x+4, 'g:')
```

```python
#Customización con Pyplot
fig, axes = plt.subplots(figsize=(4,4))

axes.plot(x, x+1, color='green', alpha=0.5) #Congigiración de color y transparencia
axes.plot(x, x+2, color='green', linewidth=8, linestyle=':', marker='x', markersize='10', markerfacecolor='#CCFFCC') # Ajuste de grosor y tipo de línea. Ajuste en estilo de marcadores
axes.plot(x, x+3, color='blue', linewidth=5, linestyle='dashed', marker='v', markersize='5')
axes.plot(x, x+4, color='gray', linewidth=3, linestyle='--', marker='p', markersize='8')
```

#Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'ITALY', 'GERMANY', 'COLOMBIA']
population = [1000, 500, 600, 400, 800]
```

```python
plt.bar(country, population) #Gráfica de barras con variables categóricas en el eje 'x'
```

```python
plt.bar(country, population, width=0.6)
plt.show()
```

```python
plt.bar(country, population, width=0.6, color='red') #Ajuste de color de barras
plt.show()
```

```python
plt.bar(country, population, width=0.6, color=['red', 'blue']) #Ajuste de color intercalado
plt.show()
```

```python
plt.bar(country, population, width=0.6, color=['red', 'blue'])
plt.xticks(np.arange(5),('INDIA', 'ITALY', 'GERMANY', 'COLOMBIA', 'JAPAN')) #Ajusta orden en un arreglo de 'np'
plt.show()
```

```python
plt.bar(country, population, width=0.6, color=['red', 'blue'])
plt.xticks(np.arange(5),('INDIA', 'ITALY', 'GERMANY', 'COLOMBIA', 'JAPAN'), rotation=45) #Ajuste de rotación
plt.show()
```

```python
plt.bar(country, population, width=0.6, color=['red', 'blue'], label='N° personas')
plt.xticks(np.arange(5),('INDIA', 'ITALY', 'GERMANY', 'COLOMBIA', 'JAPAN'), rotation=60)
plt.title('Población mundial')
plt.xlabel('Pais')
plt.ylabel('Cantidad en millones')
plt.legend()
plt.show()
```

```python
plt.barh(country, population) #Configuración horizontal
plt.show()
```

#Crear otro tipo de gráficas

##Histograma

```python
data = np.random.randint(1,50,100)
data
```

```python
print(plt.style.available)
plt.style.use('Solarize_Light2')
```

```python
plt.hist(data) #Crea histograma
plt.show()
```

```python
plt.hist(data, bins=50) #bins da el parámetro para el rango de frecuencia
plt.show()
```

```python
plt.hist(data, bins=50, width=5) #Ajustar ancho de barra
plt.show()
```

```python
plt.hist(data, bins=20, histtype='step') #Mostrar contorno
plt.show()
```

##Boxplot

```python
plt.boxplot(data) #Crear diagrama de caja
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True) #Ajuste en vertical y mejora en visualización
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True) #Ajuste en visualización
plt.show()
```

```python
data = np.append(data, 100) #Se agreaga un outlier
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True) #Visualización de outliers
plt.show()
```

##Scatterplot

```python
N = 50
x = np.random.rand(N) #Crear datos aleatorios
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2 #Ajustar dimensiones gráfico
colors = np.random.rand(N)
```

```python
plt.scatter(x,y) #Crear scatterplot
plt.show()
```

```python
plt.scatter(x,y, s=area) #Crear scatterplot y ajuste de parámetros
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors) #Crear scatterplot y ajuste de parámetros
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='x') #Crear scatterplot y ajuste de parámetros
plt.show()
```

```python
plt.scatter(x,y, s=area, c=colors, marker='x', alpha=0.5) #Personalización de puntos y transparencia
plt.show()
```

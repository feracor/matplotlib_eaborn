```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B', 'C'], y=[3,1,2]) #Visualizar diagrama de barra
plt.show()
```

```python
sns.set(style='darkgrid') #Ajuste de gr√°fica
sns.barplot(x=['A','B', 'C'], y=[3,1,2])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark', font='Ariel', font_scale=1) #Ajuste de tipo fuente
sns.barplot(x=['A','B', 'C'], y=[3,1,2])
plt.show()
```

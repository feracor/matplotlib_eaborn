#Parámetros más usados con Seaborn

```python
tips = sns.load_dataset('tips') #Importar set de datos predefinido
tips
```

```python
sns.displot(data=tips, x='total_bill') #sns crea diagrama de barras con 'total_bill' en eje 'x'
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip') #sns sugiere un gráfico de acuerdo a los datos suministrados
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip', hue='sex') #Parámetro 'hue' hace desgloce por medio de otra variable
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex', kind='kde') #Parámetro 'kind' permite escoger el tipo de gráfico
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex', kind='kde', legend=True, palette='dark', alpha=0.2) #Ajuste de leyenda, paleta de colores y transparencia
plt.show()
```

##Histograma

```python
sns.histplot(data=tips, x='tip') #Crear histograma
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=10) #Ajuste de rango con 'bins', que es equivalente a número de barras
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True) #Acumulación de datos representados gráficamente
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='frequency') #Agrupamiento por sexo, y frecuencia con hue
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability') # Probabilidad, de 0 a 1
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='percent') #Porcentaje
plt.show()
```

```python
 sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density') #Densidad
 plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='stack') #Presentar datos stackeados
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='layer') #Presentar datos layer
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge') #Presentar datos dodge
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='fill') #Presentar datos fill, de 0 a 1
plt.show()
```

#Graficar variables categóricas

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

##Countplot

```python
sns.countplot(data=tips, x='day', hue='sex') #Tomar variable categórica 'day'
plt.show()
```

```python
sns.countplot(data=tips, y='day', hue='sex') #Visualización vertical
plt.show()
```

##Stripplot

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

##Swarmplot

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) #Permite ver la concentración de datos
plt.show()
```

##Boxplot

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) #Distribucíón por percentiles
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color='0', marker='d') #Combinación con otro tipo de gráfico
plt.show()
```

##Violinplot

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, split=True, palette='pastel') #'split' permite juntar
plt.show()
```

```python
sns.violinplot(data=tips, y='tip') #Gráfica de dato cuantitativo
plt.show()
```

##Catplot

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)  #Similar a stripplot
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, col='time')  #'col' genera una comparación entre gráficos, pero diferente variable
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, col='time', kind='box')  #Cambio a boxplot con 'kind'
plt.show()
```

#Relation

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

##Scatterplot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day') #Segmentación por dia
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time') #Segmentación por estilo y hora del dia
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') #Segmentación por cantidad de comensales
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5)) #Ajustar parámetros de la leyenda por fuera del gráfico
plt.show()
```

```python
plt.figure(figsize=(8,8)) #Ajustar tamaño de gráfico
markers={'Lunch':'D', 'Dinner':'s'} #Predefinir los estilos de los markers por merdio de condicionales en un diccionario
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers) #Ajustar los markers
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

##Lineplot

```python
plt.figure(figsize=(9,7))
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

##Relplot

```python
plt.figure(figsize=(9,7))
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.show()
```

```python
plt.figure(figsize=(9,7))
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time')
plt.show()
```
